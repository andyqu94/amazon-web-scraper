```python
#create empty lists to hold cleaned value
name = []
review = []
price = []
#split the information of the product by row and append them into the empty lists
for item in product:
    clean_by_row = item.split("\n")[1]
    name.append(clean_by_row)

for item in product:
    clean_by_row1 = item.split("\n")[2]
    review.append(clean_by_row1)

for item in product:
    clean_by_row2 = item.split("\n")[-1]
    price.append(clean_by_row2)
```

```python
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float, ForeignKey
from sqlalchemy.orm import sessionmaker, relationship

# create and connect to the data base
engine = create_engine('mysql+mysqldb://root:177000@127.0.0.1:3306/web?charset=utf8', encoding = 'utf-8')
conn = engine.connect()

# declare a base object
Base = declarative_base()
```

```python
def populate_table(engine, table, df):
    """Populates a table from a Pandas DataFrame."""
    # connect to the database
    conn = engine.connect()
    
    # Orient='records' creates a list of data to write
    # http://pandas-docs.github.io/pandas-docs-travis/io.html#orient-options
    data = df.to_dict(orient='records')

    # Optional: Delete all rows in the table before insert
    conn.execute(table.delete())

    # Insert the dataframe into the database in one bulk insert
    conn.execute(table.insert(), data)
```

```python
# create table for review 
class review(Base):    
    __tablename__ = 'review'  
    name = Column(String(200), primary_key=True, nullable=True)
    reviews = Column(Integer, nullable=True)

    def __repr__(self):
        return f"id={self.name}, name={self.reviews}"
```

```python
#create table for price
class price(Base): 
    __tablename__ = 'price'
    name = Column(String(200), ForeignKey('review.name'), primary_key=True, nullable=True)
    price = Column(String(200), nullable=True)

    def __repr__(self):
        return f"id={self.name}, name={self.price}"
```

```python
# create the tables in the base
Base.metadata.create_all(engine)
engine.table_names() #check that the tables have been created
```

```python
# populate the tables in the base
populate_table(engine,review.__table__, review_df)
populate_table(engine,price.__table__, price_df)
```
